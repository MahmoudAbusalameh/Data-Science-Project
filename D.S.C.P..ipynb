## Number of Homicides on USA Analysis based on States 

```python
_ = sns.countplot(x='State', data=df_HR)
plt.xticks(rotation=90)
plt.title('Number of Homicides in USA from 1980-2014')
plt.xlabel('State')
plt.ylabel('Number of Homicides')
plt.show()
```

As we can see, California is the state with the highest incidents then comes Texas , New York and Florida.

## Analysis based on years 

```python
_ = sns.countplot(x='Year', data=df_HR)
plt.xticks(rotation=90)
plt.title('Number of Homicides in USA from 1980-2014')
plt.xlabel('Year')
plt.ylabel('Number of Homicides')
plt.show()
```

Years 1993, 1992, 1994, 1980 have the highest numbers of crimes, and we can notice  that their's a decline in the number of incidents as we move to 2014

```python
d2 = df_HR[['Perpetrator Race','Crime Solved','Year']].groupby(by=['Perpetrator Race','Crime Solved']).agg({'Year':'count'})
print(d2.groupby(level=0).apply(lambda x:100 * x/float(x.sum())))
```

```python
d4= df_HR[['Victim Race','Year','Crime Solved']]
d4=d4.groupby(by=['Victim Race', 'Year','Crime Solved']).agg({'Crime Solved':'count'})
d4= d4.groupby(level=[0,1]).apply(lambda x:100 * x/x.sum())
fig = plt.figure()
ax = fig.add_subplot(111)
d4.columns=["Solved_count"]

for name, group in d4.groupby(level=0):
    if name!="Unknown":
        group= group.reset_index().drop("Victim Race",1)
        group=group[group["Crime Solved"]=="Yes"]
        ax.plot(group["Year"], group["Solved_count"], label=name)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
plt.title('Crime solving rate by victim race')
plt.show()
```

```python
d4 = df_HR[['Victim Race','Perpetrator Race','Year']].groupby(by=['Victim Race','Perpetrator Race']).agg({'Year':'count'})
d4= d4.groupby(level=0).apply(lambda x:100 * x/float(x.sum()))

fig = plt.figure()
ax = fig.add_subplot(111)

victim = d4.index.get_level_values(0).unique()

ind = np.arange(5)
plt.xticks(ind, victim)
width = 0.35
d4 = d4.reset_index()
s2=[0,0,0,0,0]
colors=['yellow','black','red','grey','white']
i=0
for r in victim:
    s =d4[d4['Perpetrator Race']==r]["Year"]
    p= ax.bar(ind,s, width, bottom=s2, color=colors[i], label=r)
    i=i+1
    s2 = [x + y for x, y in zip(s, s2)]
    
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Victims')
plt.show()
```

```python
from IPython.display import display

df_HR['Crime Solved'].replace("No",0, inplace = True)
df_HR['Crime Solved'].replace("Yes",1, inplace = True)

df_HR.loc[df_HR['Perpetrator Count'] <=1, 'Perpetrator Count'] = 0
df_HR.loc[df_HR['Perpetrator Count'] >1, 'Perpetrator Count'] = 1

multiple_caught = pd.value_counts((df_HR['Perpetrator Count']*(df_HR['Crime Solved'])), sort = False)[1]
one_escape = pd.value_counts((df_HR['Perpetrator Count'] | df_HR['Crime Solved']),sort = False)[0]
one_caught = pd.value_counts((df_HR['Perpetrator Count'] < df_HR['Crime Solved']), sort = False)[True]
multiple_escape = pd.value_counts((df_HR['Perpetrator Count'] > df_HR['Crime Solved']), sort = False)[True]

res = np.matrix([[one_caught, multiple_caught],[one_escape,multiple_escape]])
print(pd.DataFrame(res))
```

There were roughly 194000 crimes committed by a single perpetrator, 141000 of which were solved and 54000 were not. 5100 crimes were committed by multiple perpetrators, and 11 of those were unsolved.

```python
print("Percentage of unsolved cases with one perp: %f \nPercentage of unsolved cases with multiple perps: " %((one_escape/(one_caught+one_escape))),(multiple_escape/(multiple_caught+multiple_escape)))
```
